2️⃣ Hands-On Part 1: Image Emotion Detection (10 min)

Steps:

Upload an image or use a sample image.

Detect faces using OpenCV.

Analyze emotions using DeepFace.

Show results with rectangles + dominant emotion.

Code Example:

```python
import cv2
from deepface import DeepFace

# Load image
img = cv2.imread("test.jpg")

# Load face cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

# Analyze emotions
result = DeepFace.analyze(img, actions=['emotion'], enforce_detection=False)
emotion = result[0]['dominant_emotion']

# Draw rectangles and emotion
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(img, emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)

cv2.imshow("Emotion Detection", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

3️⃣ Hands-On Part 2: Real-Time Webcam Detection (Optional / 10 min)

If students have local machines:

```python
import cv2
from deepface import DeepFace

cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    try:
        result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
        emotion = result[0]['dominant_emotion']
    except:
        emotion = "No face"

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)

    cv2.imshow("Webcam Emotion Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

✅ FULL REAL-TIME EMOTION DETECTION  

```python
import cv2   # pip install opencv-python
from deepface import DeepFace   # pip install deepface

# Load Haar Cascade
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Open camera
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # DeepFace emotion detection
    result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)

    # Convert to gray for Haar Cascade
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Loop through detected faces
    for (x, y, w, h) in faces:
        # Draw rectangle around the face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Get dominant emotion
        emotion = result[0]['dominant_emotion']

        # Draw emotion text
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1
        color = (0, 0, 255)   # Red
        thickness = 2

        cv2.putText(frame,
                    emotion,
                    (x, y - 10),
                    font,
                    font_scale,
                    color,
                    thickness,
                    cv2.LINE_AA)  # Smooth text

    # Show the result
    cv2.imshow("Emotion Detector", frame)

    # Press 'q' to exit
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
